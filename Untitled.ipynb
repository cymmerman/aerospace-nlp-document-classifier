```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import re
import string

# Download all required NLTK data
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # Lowercase
    text = text.lower()
    # Remove punctuation and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Tokenize
    tokens = word_tokenize(text)
    # Remove stopwords and lemmatize
    tokens = [lemmatizer.lemmatize(t) for t in tokens 
              if t not in stop_words and len(t) > 2]
    return ' '.join(tokens)

# Apply to all documents
df['processed_text'] = df['text'].apply(preprocess_text)

# Show before and after
print("ORIGINAL TEXT (first 200 chars):")
print(df['text'].iloc[0][:200])
print("\nPROCESSED TEXT (first 200 chars):")
print(df['processed_text'].iloc[0][:200])
print(f"\nOriginal word count: {len(df['text'].iloc[0].split())}")
print(f"Processed word count: {len(df['processed_text'].iloc[0].split())}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder

# Convert text to TF-IDF feature matrix
# TF-IDF = Term Frequency × Inverse Document Frequency
# High score = word appears often in THIS doc but rarely across ALL docs
tfidf = TfidfVectorizer(
    max_features=500,      # Top 500 most informative words
    ngram_range=(1, 2),    # Single words AND two word phrases
    min_df=1,              # Must appear in at least 1 document
    sublinear_tf=True      # Dampen effect of very frequent terms
)

X = tfidf.fit_transform(df['processed_text'])
feature_names = tfidf.get_feature_names_out()

# Encode categories as numbers
le = LabelEncoder()
y = le.fit_transform(df['category'])

print(f"TF-IDF matrix shape: {X.shape}")
print(f"Documents: {X.shape[0]}, Features: {X.shape[1]}")
print(f"Categories encoded: {dict(zip(le.classes_, range(len(le.classes_))))}")
print(f"\nTop 10 features: {feature_names[:10].tolist()}")
print(f"\nSample TF-IDF scores for document 1:")
scores = X[0].toarray()[0]
top_indices = scores.argsort()[-10:][::-1]
for idx in top_indices:
    print(f"  {feature_names[idx]}: {scores[idx]:.4f}")
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, LeaveOneOut
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# With only 25 documents use Leave One Out cross validation
# This trains on 24 docs and tests on 1, repeated 25 times
loo = LeaveOneOut()

models = {
    'Naive Bayes': MultinomialNB(alpha=0.1),
    'Linear SVM': LinearSVC(C=1.0, max_iter=1000),
    'Logistic Regression': LogisticRegression(C=1.0, max_iter=1000)
}

print("Model Comparison — Leave One Out Cross Validation")
print("=" * 50)

best_model = None
best_score = 0

for name, clf in models.items():
    scores = cross_val_score(clf, X, y, cv=loo, scoring='accuracy')
    mean_score = scores.mean()
    print(f"{name:<25} Accuracy: {mean_score:.3f} ({mean_score*100:.1f}%)")
    if mean_score > best_score:
        best_score = mean_score
        best_model = name

print(f"\nBest model: {best_model} ({best_score*100:.1f}%)")

# Train final model on all data
final_model = models[best_model]
final_model.fit(X, y)
print(f"Final model trained on all {len(df)} documents")
```

```python
# Confusion matrix — where does the model get confused?
from sklearn.model_selection import cross_val_predict

y_pred = cross_val_predict(final_model, X, y, cv=loo)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Confusion matrix
cm = confusion_matrix(y, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=le.classes_,
            yticklabels=le.classes_,
            ax=axes[0])
axes[0].set_title('Confusion Matrix — Where Does the Model Get Confused?')
axes[0].set_ylabel('Actual Category')
axes[0].set_xlabel('Predicted Category')
axes[0].tick_params(axis='x', rotation=45)

# Plot 2: Top keywords per category
category_keywords = {}
for i, category in enumerate(le.classes_):
    # Get documents in this category
    cat_indices = [j for j, label in enumerate(y) if label == i]
    cat_text = ' '.join(df['processed_text'].iloc[cat_indices])
    
    # Get TF-IDF scores for this category
    cat_vector = tfidf.transform([cat_text]).toarray()[0]
    top_idx = cat_vector.argsort()[-5:][::-1]
    category_keywords[category] = [feature_names[idx] for idx in top_idx]

# Display as table
y_pos = np.arange(len(le.classes_))
colors = ['#58a6ff', '#3fb950', '#d29922', '#f85149', '#bc8cff']
axes[1].axis('off')
table_data = [[cat] + kws for cat, kws in category_keywords.items()]
table = axes[1].table(
    cellText=table_data,
    colLabels=['Category', 'KW1', 'KW2', 'KW3', 'KW4', 'KW5'],
    cellLoc='left',
    loc='center'
)
table.auto_set_font_size(False)
table.set_fontsize(9)
table.scale(1.2, 2.0)
axes[1].set_title('Top Keywords Per Category', pad=20)

plt.tight_layout()
plt.savefig('classifier_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved classifier_analysis.png")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def classify_document(text):
    processed = preprocess_text(text)
    vector = tfidf.transform([processed])
    prediction = final_model.predict(vector)[0]
    category = le.inverse_transform([prediction])[0]
    return category

def search_documents(query, top_n=5):
    # Preprocess query
    processed_query = preprocess_text(query)
    query_vector = tfidf.transform([processed_query])
    
    # Calculate similarity to all documents
    similarities = cosine_similarity(query_vector, X)[0]
    
    # Get top results
    top_indices = similarities.argsort()[-top_n:][::-1]
    
    results = []
    for idx in top_indices:
        results.append({
            'rank': len(results) + 1,
            'title': df['title'].iloc[idx],
            'category': df['category'].iloc[idx],
            'similarity': similarities[idx]
        })
    return results

def find_similar_documents(doc_idx, top_n=3):
    doc_vector = X[doc_idx]
    similarities = cosine_similarity(doc_vector, X)[0]
    similarities[doc_idx] = 0  # exclude itself
    top_indices = similarities.argsort()[-top_n:][::-1]
    return [(df['title'].iloc[i], df['category'].iloc[i], 
             similarities[i]) for i in top_indices]

# Test the search engine
print("DOCUMENT SEARCH ENGINE")
print("=" * 60)

queries = [
    "missile guidance inertial navigation accuracy",
    "hypersonic vehicle thermal protection heating",
    "radar sensor target tracking electronic warfare"
]

for query in queries:
    print(f"\nQuery: '{query}'")
    print("-" * 60)
    results = search_documents(query, top_n=3)
    for r in results:
        print(f"  {r['rank']}. [{r['category']:>12}] {r['title'][:45]} "
              f"(similarity: {r['similarity']:.3f})")
```
